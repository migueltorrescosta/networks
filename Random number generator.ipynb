## All imports

```python
import random
from pynverse import inversefunc as inv
import math
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

## Page 5 Ex. 2

```python
# This function takes a cdf as input and outputs a random number generator with that cdf
def random_generator_functor(cdf):
    f = inv(cdf)
    def generator():
        x = random.random()
        return f(x)
    return generator
```

## Plotting script

```python
# The red dotted line is what we expect
# The blue bars is the observed values
def plot_generator(cdf,samples=1000):
    f = random_generator_functor(cdf)
    sample = [f() for i in range(samples)]
    n, bins, patches = plt.hist(sample,bins=100,density=True,cumulative=True)
    y = [cdf(x) for x in bins]
    l = plt.plot(bins, y, 'r--', linewidth=1)
    plt.show()
```

## Example with Sigmoid function

```python
Sigmoid = lambda x:1/(1+math.exp(-x))
plot_generator(cdf=Sigmoid)
```

## Example with Gaussian distribution

```python
Gaussian = lambda x:(1+math.erf(x/math.sqrt(2)))/2
plot_generator(cdf=Gaussian)
```
